```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
obs_rf2_0_m23m25 = pd.read_csv("obs_rf2_0_m23m25.csv")
# obs_lgb2_0_m23m25 = pd.read_csv("obs_lgb2_0_m23m25.csv") # p=0.35
obs_rf3_0_m23m25 = pd.read_csv("obs_rf3_0_m23m25.csv")
obs_lgb3_0_m23m25 = pd.read_csv("obs_lgb3_0_m23m25.csv")

all_obs = [
    obs_rf2_0_m23m25,
#     obs_lgb2_0_m23m25,
    obs_rf3_0_m23m25,
    obs_lgb3_0_m23m25,
]
all_labels = [
    'rf2_0_m23m25',
#     'lgb2_0_m23m25',
    'rf3_0_m23m25',
    'lgb3_0_m23m25',
]
```

```python
def add_cols(df):
    if 'date' in df.columns and 'time' in df.columns:
        df['datetime'] = (df['date'] + ' ' + df['time']).apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d %I:%M %p'))
        df.drop(['date','time'], axis=1, inplace=True)
        
    df['correct'] = (df['winner'].astype(str) == df['predicted_winner'].astype(str))*1

    def eval_odds(x):
        if isinstance(x, int) or isinstance(x, float):
            return float(x)
        if "/" in x:
            parts = x.split("/")
            return float(parts[0]) / float(parts[1])
        return float(x)
    df['odds_decimal'] = df['winner_odds'].apply(lambda x: eval_odds(x))

    def eval_return(x):
        if isinstance(x, int) or isinstance(x, float):
            return float(x) + 1
        if "/" in x:
            parts = x.split("/")
            return (float(parts[0]) + float(parts[1])) / float(parts[1])
        return float(x)
    df['returns_decimal'] = df['winner_odds'].apply(lambda x: eval_return(x))
    
    df['race_return'] = df['correct'] * df['returns_decimal']
    
for ob in all_obs:
    add_cols(ob)
```

# Simulations

```python
def size_bet_odds(pool, s, p, h0, lr=10, precision=0.1, max_iters=10000):
    b = 0
    iters = 0
    prev_step_size = 1
    
    while iters < max_iters and prev_step_size > precision:
        dr = -(s*pool*p+2*s*p*b)/(h0+b) + (s*pool*b*p+s*p*(b**2))/((h0+b)**2) + 1
        prev_b = b
        b = b - lr*dr
        prev_step_size = abs(b-prev_b)
        iters += 1
    if iters >= max_iters:
        pass
#         print('reached max')
    return b, s*(pool+b)/(h0+b)-1

def sim(df, money, p, shuffle=False):
    for i in range(df.shape[0]):
        if df['confidence'][i] >= 0.9:
            h0 = df['winner_pool_size'][i]
            s = 0.8
            pool = df['pool_size'][i]
            b, odds = size_bet_odds(pool,s,p,h0)
            bet_size = min(max(b, 0), money[-1]*0.1)
            
            if df['correct'][i]:
#                 print('1', b, odds, 'change:', bet_size*odds)
#                 print(pool, s, p, h0)
                money.append(money[-1] + bet_size * odds)
            else:
#                 print('0', b, odds, 'change:', -bet_size)
                money.append(money[-1] - bet_size)
            
        else:
            money.append(money[-1])

def sim_plot(df, p):
    df = df.sort_values('datetime')

    runs = []
    for r in range(1):
        money = [1000]
        sim(df, money, p=p)
        runs.append(np.array(money))

    plt.figure(figsize=(20,3))
    plt.ylabel('cash')
    plt.xlabel('# horse bets')
    plt.grid()
    plt.plot(df['datetime'], np.average(runs, axis=0)[1:])
    plt.show()
```

```python
sim_plot(obs_rf2_0_m23m25,p=0.365)
```

```python
sim_plot(obs_rf3_0_m23m25,p=0.25)
```

```python
sim_plot(obs_lgb3_0_m23m25, p=0.2)
```

### *** The Kelly criterion assumes that the odds don't change. But for horseracing, they do ***

Below is the graph modeling the optimal bet size (b), when odds are affected by bets.

```python
plt.plot([b*(s*(pool+b)/(h0+b) - 1)*p + (1-p)*-b for b in np.arange(0, 20001, 1000)])
plt.title('expected return vs b (thousands)')
plt.xlabel('b')
plt.ylabel('expected return')
plt.grid()
plt.show()
```

Given pool, h0, p, I can use gradient descent to find the optimal size for b that maximizes the return `(odds*b)p + (-b)(1-p)`. odds = `s*(pool+b)/(h0+b) - 1`. I'll need to hand-calculate the gradient and hardcode it for GD.

```python
def size_bet(pool, s, p, h0, lr=10, precision=0.1, max_iters=10000):
    b = 0
    iters = 0
    prev_step_size = 1
    
    while iters < max_iters and prev_step_size > precision:
        dr = -(s*pool*p+2*s*p*b)/(h0+b) + (s*pool*b*p+s*p*(b**2))/((h0+b)**2) + 1
        prev_b = b
        b = b - lr*dr
        prev_step_size = abs(b-prev_b)
        iters += 1
    print(iters)
    return b
```
