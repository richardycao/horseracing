```python
from os import listdir
from os.path import isfile, join
import pandas as pd
from collections import defaultdict, Counter
from tqdm import tqdm
from io import StringIO
from csv import writer 
```

### Notes

Find how much missing data there is for each feature. To do this, scan across 1 type of csv over all races. Get all of the features from that csv.

```python
path = '../scrape/results'
dates = [f for f in listdir(path) if not isfile(join(path, f))]

csv_presence = defaultdict(lambda: 0)
features = defaultdict(lambda: [])

for i, d in tqdm(enumerate(dates)):
    races = [f for f in listdir(f'{path}/{d}') if not isfile(join(f'{path}/{d}', f))]
    for r in races:
        csvs = [f for f in listdir(f'{path}/{d}/{r}') if isfile(join(f'{path}/{d}/{r}', f))]
        
        #######
        name = 'details'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_distance'].extend(df['distance'])
            features[f'{name}_racetype'].extend(df['race type'])
            features[f'{name}_raceclass'].extend(df['race class'])
            features[f'{name}_surfacename'].extend(df['surface name'])
            features[f'{name}_defaultcondition'].extend(df['default condition'])
            
        name = 'results'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_ranking'].extend(df['ranking'])
            features[f'{name}_horsenumber'].extend(df['horse number'])
            features[f'{name}_horsename'].extend(df['horse name'])
            features[f'{name}_win'].extend(df['win'])
            features[f'{name}_place'].extend(df['place'])
            features[f'{name}_show'].extend(df['show'])
            
        name = 'racecard_left'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_number'].extend(df['number'])
            features[f'{name}_runnerodds'].extend(df['runner odds'])
            features[f'{name}_racemorningodds'].extend(df['race morning odds'])
            features[f'{name}_name'].extend(df['name'])
            features[f'{name}_age'].extend(df['age'])
            features[f'{name}_gender'].extend(df['gender'])
            features[f'{name}_siredam'].extend(df['sire dam'])
            
        name = 'racecard_summ'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_med'].extend(df['med'])
            features[f'{name}_trainer'].extend(df['trainer'])
            features[f'{name}_weight'].extend(df['weight'])
            features[f'{name}_jockey'].extend(df['jockey'])
            
        name = 'racecard_snap'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_powerrating'].extend(df['power rating'])
            features[f'{name}_winsstarts'].extend(df['wins/starts'])
            features[f'{name}_daysoff'].extend(df['days off'])
            
        name = 'racecard_spee'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_avgspeed'].extend(df['avg speed'])
            features[f'{name}_avgdistance'].extend(df['avg distance'])
            features[f'{name}_highspeed'].extend(df['high speed'])
            features[f'{name}_avgclass'].extend(df['avg class'])
            features[f'{name}_lastclass'].extend(df['last class'])
            
        name = 'racecard_pace'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_numraces'].extend(df['num races'])
            features[f'{name}_early'].extend(df['early'])
            features[f'{name}_middle'].extend(df['middle'])
            features[f'{name}_finish'].extend(df['finish'])
            
        name = 'racecard_jock'
        if f'{name}.csv' in csvs:
            csv_presence[f'{name}'] += 1
            df = pd.read_csv(f'{path}/{d}/{r}/{name}.csv')
            features[f'{name}_starts'].extend(df['starts'])
            features[f'{name}_1st'].extend(df['1st'])
            features[f'{name}_2nd'].extend(df['2nd'])
            features[f'{name}_3rd'].extend(df['3rd'])
```

### Combined data

Now I need to combine all of these stats into 1 dataframe and see how much overlap there is with good/bad data. These will be the 1v1 samples for now.

Pairings:
* Every pair of horses in a race.
* In each pair, do both orderings.
* Each pair must have at 1 horse who placed in the results. Otherwise, we have no way of telling who won between them.

#### Do the following for each race:

Get the race time from the title of the race.

Every sample (pairing) from a race will have the same details: 'details_distance','details_racetype','details_raceclass','details_surfacename','details_defaultcondition'.

Every sample's y-value (label) is calculated from 'results_ranking' and 'results_horsenumber'. It's 0 if the 1st horse won, and 1 if the second horse won.

Exclude (for now) 'results_horsename','results_win','results_place','results_show'.

For the racecard_* tables,
* First, merge them into 1 table. This means removing the scratched ('-') from racecard_left, and then combining it with the rest of the racecard_* tables.
* After there is 1 master table, take every pairing and ordering of horses, where there is at least 1 horse who placed.
* For each horse in a pair, get
   'racecard_left_runnerodds','racecard_left_racemorningodds','racecard_left_age','racecard_left_gender',
   'racecard_summ_trainer',
    'racecard_summ_weight','racecard_summ_jockey','racecard_snap_powerrating','racecard_snap_winsstarts',
    'racecard_snap_daysoff','racecard_spee_avgspeed','racecard_spee_avgdistance','racecard_spee_highspeed',
    'racecard_spee_avgclass','racecard_spee_lastclass','racecard_pace_numraces','racecard_pace_early',
    'racecard_pace_middle','racecard_pace_finish','racecard_jock_starts','racecard_jock_1st','racecard_jock_2nd',
    'racecard_jock_3rd'
    
It comes out to be 51 features per sample. It's a lot to start with, but I can probably remove some later.



```python
# For reference

default_columns=[
    'details_distance','details_racetype','details_raceclass','details_surfacename','details_defaultcondition',
    'results_ranking','results_horsenumber','results_horsename','results_win','results_place','results_show',
    'racecard_left_number','racecard_left_runnerodds','racecard_left_racemorningodds','racecard_left_name',
    'racecard_left_age','racecard_left_gender','racecard_left_siredam','racecard_summ_trainer',
    'racecard_summ_weight','racecard_summ_jockey','racecard_snap_powerrating','racecard_snap_winsstarts',
    'racecard_snap_daysoff','racecard_spee_avgspeed','racecard_spee_avgdistance','racecard_spee_highspeed',
    'racecard_spee_avgclass','racecard_spee_lastclass','racecard_pace_numraces','racecard_pace_early',
    'racecard_pace_middle','racecard_pace_finish','racecard_jock_starts','racecard_jock_1st','racecard_jock_2nd',
    'racecard_jock_3rd']
```

## Create dataframe

```python
per_horse_columns = [
    'horse_number','runner_odds','morning_odds','horse_name','horse_age','horse_gender','horse_siredam','trainer',
    'horse_med','horse_weight','jockey','horse_power_rating','horse_wins/starts','horse_days_off','horse_avg_speed',
    'horse_avg_distance','horse_high_speed','horse_avg_class','horse_last_class','horse_num_races','early','middle',
    'finish','jockey_trainer_starts','jockey_trainer_1st','jockey_trainer_2nd','jockey_trainer_3rd'
]

horse1_cols = [c+'_1' for c in per_horse_columns]
horse2_cols = [c+'_2' for c in per_horse_columns]

csv_columns_row = [
    'date','time','race_distance','race_type','race_class','surface_name','default_condition',
    *horse1_cols, *horse2_cols, 'result'
]
print(csv_columns_row)

# df = pd.DataFrame(columns=[
#     'date','time','race_distance','race_type','race_class','surface_name','default_condition',
#     *horse1_cols, *horse2_cols, 'result'
# ])
```

```python
path = '../scrape/results'
dates = [f for f in listdir(path) if not isfile(join(path, f))]
output = StringIO()
csv_writer = writer(output)

csv_writer.writerow(csv_columns_row)

races = []
for d in dates:
    races.extend([f'{path}/{d}/{r}' for r in listdir(f'{path}/{d}') if not isfile(join(f'{path}/{d}', r))])

for r in tqdm(races):
    csvs = [c for c in listdir(r) if isfile(join(r, c))]

    # Skips races with missing files
    missing_csv = False
    for n in ['details','results','racecard_left','racecard_summ','racecard_snap','racecard_spee',
              'racecard_pace','racecard_jock']:
        if f'{n}.csv' not in csvs:
            missing_csv = True
    if missing_csv:
        continue

    # Get race time
    race_parts = r.split('/')[-1].split('_')
    datetime_row = [r.split('/')[3]] + [' '.join([p for p in reversed(race_parts[:2])])]

    # Get details
    details = pd.read_csv(f'{r}/details.csv')
    details_row = details.iloc[0,1:].values.flatten()

    # Get results
    results = pd.read_csv(f'{r}/results.csv')
    ranked_horse_numbers = set(results['horse number'].values)

    # Racecard
    # 1. Get the list of horses from racecard_left_number.
    # 2. Iterate through all pairs of horses.
    # 3. For each pair, check that at least 1 horse is present in results_horsenumber.
    # 4. For each horse in the pair, get all the values in that list and put it into a row.
    # 5. For each pair, combine race_time_row, details_row, and pair_row into 1 sample in df.
    racecard_left = pd.read_csv(f'{r}/racecard_left.csv')
    racecard_left = racecard_left[racecard_left['runner odds'] != '-']
    num_horses = racecard_left.shape[0]
    # for horse i and horse j
    for i in range(num_horses):
        for j in range(num_horses):
            i_num = racecard_left.iloc[i]['number']
            j_num = racecard_left.iloc[j]['number']
            if i != j and (i_num in ranked_horse_numbers or 
                           j_num in ranked_horse_numbers):
                racecard_summ = pd.read_csv(f'{r}/racecard_summ.csv')
                racecard_snap = pd.read_csv(f'{r}/racecard_snap.csv')
                racecard_spee = pd.read_csv(f'{r}/racecard_spee.csv')
                racecard_pace = pd.read_csv(f'{r}/racecard_pace.csv')
                racecard_jock = pd.read_csv(f'{r}/racecard_jock.csv')

                i_row = pd.concat([racecard_left.iloc[i,1:], racecard_summ.iloc[i,1:], racecard_snap.iloc[i,1:], racecard_spee.iloc[i,1:], racecard_pace.iloc[i,1:], racecard_jock.iloc[i,1:]])
                j_row = pd.concat([racecard_left.iloc[j,1:], racecard_summ.iloc[j,1:], racecard_snap.iloc[j,1:], racecard_spee.iloc[j,1:], racecard_pace.iloc[j,1:], racecard_jock.iloc[j,1:]])

                i_rank = results.loc[results['horse number'] == i_num]['ranking'].values[0] if i_num in ranked_horse_numbers else None
                j_rank = results.loc[results['horse number'] == j_num]['ranking'].values[0] if j_num in ranked_horse_numbers else None

                y_row = -1
                if i_rank == None and j_rank != None:
                    y_row = 1
                elif i_rank != None and j_rank == None:
                    y_row = 0
                elif i_rank != None and j_rank != None:
                    y_row = 0 if i_rank < j_rank else 1

                row = [*datetime_row, *details_row, *i_row, *j_row, *[y_row]]
                csv_writer.writerow(row)
output.seek(0)
df = pd.read_csv(output)
```
